```python
# your code here
fig, ax = plt.subplots(1,2,figsize = (10,5))

ax[0].plot(x,x**2, label="x")
ax[0].set_title('Normal scale plot', fontsize=15)
ax[1].set_yscale('linear')
ax[0].legend(["aaaaa"])

ax[1].plot(x,np.exp(x), label='$y=e^x$')
ax[1].set_title('Logarithmic scale', fontsize=15)
ax[1].set_yscale('log')
ax[1].legend(["bbbbb"])
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
plt.figure(figsize=(12,4))

fitbit_grouped=fitbit.groupby(['Months_encoded']).sum()
plt.subplot(1,2,1)
plt.plot(fitbit_grouped.index, fitbit_grouped["Steps"].values)
plt.title("Steps per month on fitbit app")
plt.xlabel("Months")
plt.ylabel("Steps")
plt.xticks(np.arange(1,13))


# MEDIAN STEPS BY WORK_OR_WEEKEND
fitbit_grouped2=fitbit.groupby(['Work_or_Weekend']).sum()
plt.subplot(1,2,2)
plt.bar(fitbit_grouped2.index, fitbit_grouped2["Steps"].values)
plt.xticks(fitbit_grouped2.index, ['Weekdays', 'Weekends'])
```

#### Write a loop to plot 3 scatter plots of the following features:
